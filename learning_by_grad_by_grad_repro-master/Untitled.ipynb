```python
import seaborn as sns
sns.set(color_codes=True)
sns.set_style("white")
import numpy as np
import matplotlib.pyplot as plt
```

```python
fit_data = np.load('../mnist-data-fit.npy')
```

```python
OPT_NAMES = ['ADAM', 'RMSprop', 'SGD', 'NAG']
ax = sns.tsplot(data=fit_data[:, :, :], condition=OPT_NAMES + ['LSTM'], linestyle='--', color=['r', 'b', 'g', 'k', 'y'])
ax.lines[-1].set_linestyle('-')
ax.legend()
plt.yscale('log')
plt.xlabel('steps')
plt.ylabel('loss')
plt.title('MNIST (2 layers) [our version]')
plt.savefig('mnist2layers')
# plt.show()
```
